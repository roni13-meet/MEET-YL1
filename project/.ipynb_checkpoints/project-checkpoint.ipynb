```python
import csv
from pylab import *

f = open('animal.csv', 'rb')
reader = csv.reader(f)
bodyweight = []
brainweight = []
rownum = 0

for row in reader:
    if rownum > 0:
        b = map(float, row[1:])
    bodyweight.append(b[0])
    brainweight.append(b[1])
    rownum += 1
    

f.close()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('body weight Vs. brain weight of animals')
ax.set_xlabel('body weight')
ax.set_ylabel('brain weight')
scatter(bodyweight, brainweight)
```

As you can see, there is no relashionship between the animal's body weight and the animal's brain weight.
