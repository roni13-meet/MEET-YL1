```python
import csv
from pylab import *

f = open('cig.csv', 'rb')
reader = csv.reader(f)
tar = []
nicotine = []

for row in reader:
    a = row[0].split()
    b = map(float, a[1:])
    nicotine.append(b[0])
    tar.append(b[1])
    
print nicotine
print tar
f.close()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('cigarettes')
ax.set_xlabel('tar')
ax.set_ylabel('nicotine')
ax.scatter(tar, nicotine)
```
